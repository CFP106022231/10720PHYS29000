```python
import matplotlib.pyplot as plt
import numpy as np
import math
import random
plt.xlabel("distance(m)")
plt.ylabel("height(m)")
plt.title("angry birds")
v0=100
theta=45
g=9.8
a = 2*v0*math.sin(theta)/g
b = 2*v0*math.sin(theta)/g*v0*math.cos(theta)
t = np.arange(0,a,0.01)
x = v0*(np.cos(theta))*t
y = v0*(np.sin(theta))*t+1/2*(-g)*t**2
aa = int(a)
bb = int(b)
px = random.randint(0,bb)
py = random.randint(0,aa)
plt.plot(x,y)  
plt.plot(px,py,"o")
print(px,py)
plt.show()
if px in x and py in y:
   print("打到豬ㄌ")
else:
    print("豬豬笑你哈哈哈哈ㄍㄡˊㄍㄡˊ哈哈")
```
